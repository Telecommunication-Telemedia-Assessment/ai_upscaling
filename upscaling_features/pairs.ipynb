```python
import pandas as pd
from IPython.display import display as d
from IPython.display import Image

df = pd.read_csv("../evaluation/mos.csv")
df["dis_image"] = df["stimuli_file"].apply(
    lambda x: x.replace("./stimuli/", "../upscaling/")
)
df["src_image"] = df["base_img"].apply(
    lambda x: "../upscaling/src_images_1080/x1/" + x + ".png"
)
d(df)

Image(filename=df["dis_image"].values[2]) 

df[df["src_image"] != df["dis_image"]][["src_image", "dis_image"]].to_csv("dis_ref_pairs.csv", index=False)
```
